```python
import pandas as pd
import numpy as np

# Check what you have
print(df.shape)
print(df.columns.tolist())
print(df.isnull().sum())
print(df['track_genre'].value_counts().head(20))
print(df[['danceability','energy','valence','tempo','loudness']].describe())
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['track_name', 'artists', 'album_name'])
```

```python
df.isna().sum()
```

```python
df.head(3)
```

```python
# Drop duplicate track_ids (same song appears in multiple genre buckets — keep first)
before = len(df)
df = df.drop_duplicates(subset=['track_id'], keep='first')
print(f"Removed {before - len(df)} duplicates")
```

```python
df = df[df['tempo'] > 0]
```

```python
df = df[df['popularity'] > 0]
```

```python
df = df.reset_index(drop=True)
print(f"Clean dataset: {df.shape}")
```

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
import pickle
import numpy as np
```

```python
RL_FEATURES = ['danceability', 'energy', 'valence',
               'tempo', 'loudness', 'acousticness',
               'instrumentalness', 'speechiness']
```

```python
scaler = MinMaxScaler()
feature_matrix = scaler.fit_transform(df[RL_FEATURES])
```

```python
print("Feature matrix shape:", feature_matrix.shape)
print("Min values (should all be 0):", feature_matrix.min(axis=0))
print("Max values (should all be 1):", feature_matrix.max(axis=0))
print("\nSample row 0 (normalized):", feature_matrix[0])
print("Sample song:", df.iloc[0]['track_name'], "by", df.iloc[0]['artists'])

```

```python
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print("\n✅ Scaler saved")
```
